```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use(['classic', 'ggplot'])
```

```python
C = pd.read_csv('C_cols.csv') 
ifd = pd.read_csv('N_cols.csv')
```

```python
C = C.merge(ifd[['isFraud']], left_index=True, right_index=True)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
def plot_box(df, feature):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    sns.boxplot(x='isFraud', y=feature, data=df, ax=axes[0])
    axes[0].set_title(f'Boxplot of {feature} by isFraud')
    
    sns.boxplot(x='isFraud', y=feature, data=df, ax=axes[1])
    sns.stripplot(x='isFraud', y=feature, data=df, color='yellow', size=4, jitter=0.2, ax=axes[1])
    axes[1].set_title(f'Boxplot with IDs plot of {feature} by isFraud')
    
    sns.distplot(df[df['isFraud'] == 0][feature], kde=True, color='blue', label='Not Fraud', ax=axes[2])
    sns.distplot(df[df['isFraud'] == 1][feature], kde=True, color='red', label='Fraud', ax=axes[2])
    axes[2].set_title(f'Histogram of {feature} by isFraud')
    axes[2].legend()
    
    plt.tight_layout()
    plt.show()
```
